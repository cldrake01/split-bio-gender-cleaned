```python
# Prepare the data
age_ranges = ['0-12', '13-19', '20-29', '30-49', '50-64', '65-100']
classes = ['0-12', '13-19', '20-29', '30-49', '50-64', '65-100']

female_counts = [403, 1_840, 14_134, 19_273, 2_702, 692]
male_counts = [283, 1_053, 7_601, 23_323, 5_757, 1_525]

# Plot the bar graph
bar_width = 0.35
x_positions = np.arange(len(age_ranges))

fig, ax = plt.subplots()

# Plot the female bars
ax.bar(x_positions, female_counts, bar_width, label='Female', color='pink')

# Plot the male bars
ax.bar(x_positions + bar_width, male_counts, bar_width, label='Male', color='cyan')

ax.set_xticks(x_positions + bar_width / 2)
ax.set_xticklabels(age_ranges)
ax.set_ylabel('Quantity')
ax.set_title('Classes by Age')
ax.legend()

plt.show()
```
